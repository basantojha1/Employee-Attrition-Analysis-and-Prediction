```python
# Checking for outliers using IQR method
# Creating a numeric column
numeric_column = ['Age', 'DailyRate', 'DistanceFromHome', 'Education', 'EmployeeCount', 'EmployeeNumber', 'EnvironmentSatisfaction', 
             'HourlyRate', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 
             'PercentSalaryHike','PerformanceRating', 'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel', 
             'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 
             'YearsSinceLastPromotion', 'YearsWithCurrManager']

# Calculating Q1 and Q3
Q1 = df[numeric_column].quantile(0.25)
Q3 = df[numeric_column].quantile(0.75)

# Calculating InterQuartile
IQR = Q3 - Q1

#Setting upper and lower bounds
lower_bound = (Q1 - 1.5 * IQR)
upper_bound = (Q3 + 1.5 * IQR)

# Checking Outliers
outliers = ((df[numeric_column] < lower_bound) | (df[numeric_column] > upper_bound)).any(axis=1)
df[outliers]
```

# Bivariate Analysis

```python
# Plotting a linear line across a scatter plot
# Importing sklearn linear regression model 
from sklearn.linear_model import LinearRegression

# Step 1: Fit a linear regression model
x = df[['Age']]
y = df[['MonthlyIncome']]

model = LinearRegression()
model.fit(x, y)

# Step 2: Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['MonthlyIncome'], color='blue', alpha=0.5, label='Data Points')

# Step 3: Plot the linear line
plt.plot(x, model.predict(x), color='blue', label='Linear Regression')

plt.title('Scatter Plot of Age vs Monthly Income with Linear Regression')
plt.xlabel('Age')
plt.ylabel('Monthly Income')
plt.grid(True)
plt.legend()
plt.show()
```

# Feature engineering

```python
df.head()
```

```python
# Encoding values for Attrition (Yes, No) to (1, 0)
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
df['Attrition'].unique()
```

```python
# Gender Column unique values
df['Gender'].unique()
```

```python
# Performing one-hot encoding on the 'Gender' column
gender_encoded = pd.get_dummies(df['Gender'], prefix='Gender')

# Concatenating the encoded columns with the DataFrame
df_encoded = pd.concat([df, gender_encoded], axis=1)

# Dropping the original 'Gender' column
df_encoded.drop('Gender', axis=1, inplace=True)
```

```python
# MaritalStatus Column unique values
df['MaritalStatus'].unique()
```

```python
# Performing one-hot encoding on the 'MaritalStatus' column
marital_status_encoded = pd.get_dummies(df['MaritalStatus'], prefix='MaritalStatus')

# Concatenating the encoded columns with the DataFrame
df_encoded = pd.concat([df_encoded, marital_status_encoded], axis=1)

# Drop the original 'MaritalStatus' column
df_encoded.drop('MaritalStatus', axis=1, inplace=True)
```

```python
# Encoding values for Over18 'Yes' to '1'
df_encoded['Over18'] = df_encoded['Over18'].map({'Y': 1})
```

```python
# OverTime Column unique values
df_encoded['OverTime'].unique()
```

```python
# Encoding values for OverTime ('Yes', 'No') to (1, 0)
df_encoded['OverTime'] = df_encoded['OverTime'].map({'Yes': 1, 'No': 0})
```

```python
# BusinessTravel Column unique values
df_encoded['BusinessTravel'].unique()
```

```python
# Performing one-hot encoding on the 'BusinessTravel' column
business_travel_encoded = pd.get_dummies(df['BusinessTravel'], prefix='BusinessTravel')

# Concatenating the encoded columns with the DataFrame
df_encoded = pd.concat([df_encoded, business_travel_encoded], axis=1)

# Drop the original 'BusinessTravel' column
df_encoded.drop('BusinessTravel', axis=1, inplace=True)
```

```python
# Department Column unique values
df_encoded['Department'].unique()
```

```python
# Performing one-hot encoding on the 'Department' column
department_encoded = pd.get_dummies(df['Department'], prefix='Department')

# Concatenating the encoded columns with the DataFrame
df_encoded = pd.concat([df_encoded, department_encoded], axis=1)

# Drop the original 'Department' column
df_encoded.drop('Department', axis=1, inplace=True)
```

```python
# EducationField Column unique values
df_encoded['EducationField'].unique()
```

```python
# Performing one-hot encoding on the 'EducationField' column
education_field_encoded = pd.get_dummies(df['EducationField'], prefix='EducationField')

# Concatenating the encoded columns with the DataFrame
df_encoded = pd.concat([df_encoded, education_field_encoded], axis=1)

# Drop the original 'EduactionField' column
df_encoded.drop('EducationField', axis=1, inplace=True)
```

```python
# Performing one-hot encoding on the 'JobRole' column
job_role_encoded = pd.get_dummies(df['JobRole'], prefix='JobRole')

# Concatenating the encoded columns with the DataFrame
df_encoded = pd.concat([df_encoded, job_role_encoded], axis=1)

# Drop the original 'JobRole' column
df_encoded.drop('JobRole', axis=1, inplace=True)
```

# Data Modeling

```python
# Trying out Logistic Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Splitting data into features and target variable
x = df_encoded.drop('Attrition', axis=1)  # Features
y = df_encoded['Attrition']  # Target variable

# Splitting data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initializing the logistic regression model
logreg = LogisticRegression(max_iter=1000)  # Can adjust maximum itteration

# Fitting the model on the training data
logreg.fit(x_train, y_train)

# Maing predictions on the test set
y_predict = logreg.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_predict)
print("Accuracy of Logistic Regression Model:", accuracy)
```

```python
# Trying out Decision Tree
# Importing Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Instantiating the DecisionTreeClassifier
tree_model = DecisionTreeClassifier(random_state=42)

# Training the model
tree_model.fit(x_train, y_train)

# Predicting on the test data
y_predict_tree = tree_model.predict(x_test)

# Calculate accuracy
accuracy_tree = accuracy_score(y_test, y_predict_tree)
print("Decision Tree Model Accuracy:", accuracy_tree)
```

```python
# Importing random forest libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Splitting the data into features (x) and target variable (y)
x = df_encoded.drop(columns=['Attrition'])  # Features
y = df_encoded['Attrition']  # Target variable

# Splitting the data into training and testing sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initializing the Random Forest classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Training the classifier on the training data
rf_classifier.fit(x_train, y_train)

# Predicting the target variable on the testing data
y_pred = rf_classifier.predict(x_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Generating classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Generating confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
# Importing naive bayes libraries
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Splitting the data into features (x) and target variable (y)
x = df_encoded.drop(columns=['Attrition'])  # Features
y = df_encoded['Attrition']  # Target variable

# Splitting the data into training and testing sets (80% train, 20% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initializing the Naive Bayes classifier (GaussianNB for numerical features)
nb_classifier = GaussianNB()

# Training the classifier on the training data
nb_classifier.fit(x_train, y_train)

# Predicting the target variable on the testing data
y_predict = nb_classifier.predict(x_test)

# Evaluating the model
accuracy = accuracy_score(y_test, y_predict)
print("Accuracy:", accuracy)

# Generating classification report
print("\nClassification Report:")
print(classification_report(y_test, y_predict))

# Generating confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_predict))
```
